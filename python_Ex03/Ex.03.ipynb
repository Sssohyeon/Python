```python
count0fMask = int(input("마스크 개수 입력>>"))
eightbox = count0fMask//8
#8개들이 포장 후에 5개 들이 포장지로 포장한 다음 남은 마스크가 1개라도 있다면
if ((count0fMask%8) % 5) >0:
    fivebox = ((count0fMask%8)//5)+1 #5개들이 포장지는 1개가 추가로 더 필요함
#8개들이,5개들이 포장지로 포장한 후 남은 마스크가 하나도 없을 경우
else:
    fivebox = ((count0fMask%8)//5)




print("8개들이 포장지 개수:",eightbox)
print("5개들이 포장지 개수:",fivebox)
```
